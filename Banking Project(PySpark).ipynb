```python
def check_missing_percent(df,lst_cl):      #lst_cl = ["id","member_id",...]
    global missing_values_less_75
    global missing_values_more_75
    missing_values_less_75 = {}
    missing_values_more_75 = {}
    b = df.count()
    for i in lst_cl:
        a = df.filter(col(i).isNull()).count()
        c = (a/b) * 100
        if c >= 75:
            missing_values_more_75[i] = c
        else:    
            missing_values_less_75[i] = c
    return ({"missing values more than 75%" : missing_values_more_75, "missing values less than 75%" : missing_values_less_75}) 
```

```python
#to check if data is having duplicates or not
def check_duplicates(df,cl):
    a = df.select(cl).distinct().count()
    b = df.count()
    if a == b:
        return "No Duplicates"
    else:
        c = b - a
        print("There are",c,"Duplicates")
```

```python
def check_uniqueness(df,lst_cl):
    results = {}
    for i in lst_cl:
        a = df.select(i).filter(col(i).isNotNull()).distinct().count()
        if a == 1:
            results[i] = "Single unique value"
        else:
            results[i] = "Multiple value"
    return results
```
